# 2. Perspectiva de Producto

## 2.1. Distribución por Tipo de Propiedad: Conteo de registros por Tipo (Casa, Departamento, Suite, Local Comercial, etc).


```python
# Función para extraer el tipo de propiedad del título
def extract_property_type(title):
    if isinstance(title, str):
        title_lower = title.lower()
        if 'suite' in title_lower:
            return 'Suite'
        elif 'departamento' in title_lower:
            return 'Departamento'
        elif 'casa' in title_lower:
            return 'Casa'
        elif 'local comercial' in title_lower or 'local' in title_lower and 'comercial' in title_lower:
            return 'Local Comercial'
        elif 'oficina' in title_lower:
            return 'Oficina'
        elif 'terreno' in title_lower:
            return 'Terreno'
    return 'Otro'

# Aplicar la función para crear la nueva columna 'Tipo de Propiedad'
df_principal['Tipo de Propiedad'] = df_principal['Titulo'].apply(extract_property_type)
df_principal.head()
```

```python
# Contar la distribución de propiedades por tipo
distribucion_tipo_propiedad = df_principal['Tipo de Propiedad'].value_counts()
print("Distribución por Tipo de Propiedad")
distribucion_tipo_propiedad
```

# 2.3.  Premium por Habitación Adicional: Diferencia del precio promedio entre propiedades de 1 habitación vs. 2 habitaciones; 2 habitaciones vs. 3, 3 vs. 4, etc

```python
import pandas as pd
# Eliminar filas donde 'Num. dormitorios' o 'Precio' son NaN
df_bedrooms = df_principal.dropna(subset=['Num. dormitorios', 'Precio'])
# Convertir 'Num. dormitorios' a entero
df_bedrooms['Num. dormitorios'] = df_bedrooms['Num. dormitorios'].astype(int)
# Calcular el precio promedio por número de dormitorios
precio_promedio_por_dormitorio = df_bedrooms.groupby('Num. dormitorios')['Precio'].mean().sort_index()
print("Precio Promedio por Número de Dormitorios:")
precio_promedio_por_dormitorio
```

```python
# Calcular el premium por habitación adicional
premium_por_habitacion = {}
for i in range(1, len(precio_promedio_por_dormitorio)):
    dormitorios_actual = precio_promedio_por_dormitorio.index[i]
    dormitorios_anterior = precio_promedio_por_dormitorio.index[i-1]
    if dormitorios_actual == dormitorios_anterior + 1:
        premium = precio_promedio_por_dormitorio.iloc[i] - precio_promedio_por_dormitorio.iloc[i-1]
        premium_por_habitacion[f'{dormitorios_anterior} a {dormitorios_actual} dormitorios'] = premium
print("\nPremium por Habitación Adicional (Diferencia de Precio Promedio):")
for key, value in premium_por_habitacion.items():
    print(f"{key}: ${value:.2f}")
```

# 4. Segmentación de Mercado

```python
# Calcular los cuartiles de la columna Precio
Q1 = df_principal['Precio'].quantile(0.25)
Q3 = df_principal['Precio'].quantile(0.75)

# Categorizar
def categorizar(precio):
    if precio < Q1:
        return 'Económico'
    elif precio > Q3:
        return 'Lujo'
    else:
        return 'Medio'

df_principal['Tipo de Inmueble'] = df_principal['Precio'].apply(categorizar)
df_principal
```
